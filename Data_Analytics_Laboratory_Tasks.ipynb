1. Compute **sum of digits** of an integer (no string conversion).


```python
def sum_digits_no_string(n):
    if n < 0:
        n = abs(n) # Handle negative numbers by taking absolute value

    total_sum = 0
    while n > 0:
        digit = n % 10  # Get the last digit
        total_sum += digit  # Add it to the sum
        n //= 10  # Remove the last digit (integer division)
    return total_sum

# Test cases
number1 = 12345
print(f"The sum of digits of {number1} is: {sum_digits_no_string(number1)}")

number2 = 987
print(f"The sum of digits of {number2} is: {sum_digits_no_string(number2)}")

number3 = 0
print(f"The sum of digits of {number3} is: {sum_digits_no_string(number3)}")

number4 = -567
print(f"The sum of digits of {number4} is: {sum_digits_no_string(number4)}")
```

### How the code works:

1.  **`sum_digits_no_string(n)` function:**
    *   It takes an integer `n` as input.
    *   It handles negative numbers by converting them to their absolute value (`abs(n)`) because the sum of digits is usually considered for the magnitude of the number.
    *   `total_sum` is initialized to `0` to store the sum of the digits.

2.  **`while n > 0:` loop:**
    *   The loop continues as long as `n` is greater than `0`.
    *   **`digit = n % 10`**: The modulo operator (`%`) gives the remainder when `n` is divided by `10`, which is always the last digit of the number.
    *   **`total_sum += digit`**: This last digit is added to `total_sum`.
    *   **`n //= 10`**: The floor division operator (`//`) divides `n` by `10` and discards the remainder, effectively removing the last digit from the number. For example, if `n` was `123`, it becomes `12`.

3.  **`return total_sum`**: Once `n` becomes `0` (meaning all digits have been processed), the function returns the accumulated `total_sum`.

2. Compute **product of digits** (no string).

```python
def product_digits_no_string(n):
    if n == 0: # Product of digits of 0 is 0
        return 0
    if n < 0:
        n = abs(n) # Handle negative numbers by taking absolute value

    product = 1
    while n > 0:
        digit = n % 10  # Get the last digit
        product *= digit  # Multiply it to the product
        n //= 10  # Remove the last digit (integer division)
    return product

# Test cases
number1 = 12345
print(f"The product of digits of {number1} is: {product_digits_no_string(number1)}")

number2 = 987
print(f"The product of digits of {number2} is: {product_digits_no_string(number2)}")

number3 = 0
print(f"The product of digits of {number3} is: {product_digits_no_string(number3)}")

number4 = -567
print(f"The product of digits of {number4} is: {product_digits_no_string(number4)}")
number5 = 105
print(f"The product of digits of {number5} is: {product_digits_no_string(number5)}")
```

```python
def reverse_number_no_string(n):
    # Helper function to reverse a number, as developed previously
    is_negative = False
    if n < 0:
        is_negative = True
        n = abs(n)

    reversed_num = 0
    temp_n = n # Use a temporary variable to preserve original 'n' for the loop
    while temp_n > 0:
        digit = temp_n % 10
        reversed_num = reversed_num * 10 + digit
        temp_n //= 10

    return -reversed_num if is_negative else reversed_num

def is_palindrome_number_no_string(n):
    if n < 0: # Negative numbers are generally not considered palindromes
        return False
    if n == 0: # 0 is a palindrome
        return True

    # A number is a palindrome if it's equal to its reverse
    return n == reverse_number_no_string(n)

# Test cases
number1 = 121
print(f"Is {number1} a palindrome? {is_palindrome_number_no_string(number1)}")

number2 = 12321
print(f"Is {number2} a palindrome? {is_palindrome_number_no_string(number2)}")

number3 = 12345
print(f"Is {number3} a palindrome? {is_palindrome_number_no_string(number3)}")

number4 = 0
print(f"Is {number4} a palindrome? {is_palindrome_number_no_string(number4)}")

number5 = -121
print(f"Is {number5} a palindrome? {is_palindrome_number_no_string(number5)}")

number6 = 1001
print(f"Is {number6} a palindrome? {is_palindrome_number_no_string(number6)}")

number7 = 10
print(f"Is {number7} a palindrome? {is_palindrome_number_no_string(number7)}")
```

5. Check **Armstrong number** (general n-digit).

```python
def is_armstrong_number(n):
    if n < 0:
        return False # Armstrong numbers are typically positive
    if n == 0:
        return True # 0 is considered an Armstrong number

    original_n = n
    num_digits = 0
    temp_n = n
    # Count the number of digits
    while temp_n > 0:
        temp_n //= 10
        num_digits += 1

    sum_of_powers = 0
    temp_n = n # Reset temp_n to the original number for digit extraction
    # Calculate the sum of digits raised to the power of num_digits
    while temp_n > 0:
        digit = temp_n % 10
        sum_of_powers += digit ** num_digits
        temp_n //= 10

    return sum_of_powers == original_n

# Test cases
number1 = 153 # 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153
print(f"Is {number1} an Armstrong number? {is_armstrong_number(number1)}")

number2 = 370 # 3^3 + 7^3 + 0^3 = 27 + 343 + 0 = 370
print(f"Is {number2} an Armstrong number? {is_armstrong_number(number2)}")

number3 = 9474 # 9^4 + 4^4 + 7^4 + 4^4 = 6561 + 256 + 2401 + 256 = 9474
print(f"Is {number3} an Armstrong number? {is_armstrong_number(number3)}")

number4 = 123
print(f"Is {number4} an Armstrong number? {is_armstrong_number(number4)}")

number5 = 0
print(f"Is {number5} an Armstrong number? {is_armstrong_number(number5)}")

number6 = -153
print(f"Is {number6} an Armstrong number? {is_armstrong_number(number6)}")
```

### How the code works:

1.  **`is_armstrong_number(n)` function:**
    *   It takes an integer `n` as input.
    *   **Edge Cases:**
        *   Negative numbers are generally not considered Armstrong numbers, so it returns `False` for `n < 0`.
        *   `0` is considered an Armstrong number, so it returns `True` if `n == 0`.
    *   `original_n` stores the initial value of `n` for comparison later.
    *   `num_digits` is initialized to `0` to count the number of digits.
    *   `temp_n` is used as a temporary variable to count digits without modifying `original_n`.

2.  **Counting Digits:**
    *   A `while temp_n > 0:` loop repeatedly performs floor division by 10 (`temp_n //= 10`) until `temp_n` becomes `0`. In each iteration, `num_digits` is incremented, effectively counting the digits.

3.  **Calculating Sum of Powers:**
    *   `sum_of_powers` is initialized to `0`.
    *   `temp_n` is reset to `original_n` to extract digits again.
    *   Another `while temp_n > 0:` loop extracts each `digit` using the modulo operator (`% 10`).
    *   The extracted `digit` is then raised to the power of `num_digits` (`digit ** num_digits`) and added to `sum_of_powers`.
    *   `temp_n` is again floor-divided by 10 to remove the last digit.

4.  **Final Check:**
    *   Finally, the function returns `True` if `sum_of_powers` is equal to `original_n`, indicating it's an Armstrong number; otherwise, it returns `False`.

6. Check **Perfect number**.

```python
def is_perfect_number(n):
    if n <= 1: # Perfect numbers are positive integers greater than 1
        return False

    sum_divisors = 1 # Start with 1, as 1 is always a divisor
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            sum_divisors += i
            if i != n // i: # Add the pair divisor if it's not the same as i
                sum_divisors += n // i

    return sum_divisors == n

# Test cases
number1 = 6 # Divisors: 1, 2, 3. Sum = 1+2+3 = 6
print(f"Is {number1} a perfect number? {is_perfect_number(number1)}")

number2 = 28 # Divisors: 1, 2, 4, 7, 14. Sum = 1+2+4+7+14 = 28
print(f"Is {number2} a perfect number? {is_perfect_number(number2)}")

number3 = 496 # Divisors: 1, 2, 4, 8, 16, 31, 62, 124, 248. Sum = 496
print(f"Is {number3} a perfect number? {is_perfect_number(number3)}")

number4 = 8128 # Divisors: 1, 2, 4, 8, 16, 32, 64, 127, 254, 508, 1016, 2032, 4064. Sum = 8128
print(f"Is {number4} a perfect number? {is_perfect_number(number4)}")

number5 = 10
print(f"Is {number5} a perfect number? {is_perfect_number(number5)}")

number6 = 1
print(f"Is {number6} a perfect number? {is_perfect_number(number6)}")

number7 = 0
print(f"Is {number7} a perfect number? {is_perfect_number(number7)}")
```

### How the code works:

1.  **`is_perfect_number(n)` function:**
    *   It takes an integer `n` as input.
    *   **Edge Cases:**
        *   Perfect numbers are positive integers greater than 1, so it returns `False` if `n <= 1`.
    *   `sum_divisors` is initialized to `1` because `1` is always a proper divisor of any positive integer.

2.  **Finding Divisors:**
    *   The code iterates from `2` up to the square root of `n` (`int(n**0.5) + 1`). This is an optimization because divisors always come in pairs (except for the square root of perfect squares, which isn't applicable here for proper divisors).
    *   If `i` is a divisor of `n` (`n % i == 0`):
        *   `i` is added to `sum_divisors`.
        *   The complementary divisor (`n // i`) is also a divisor. If `i` is not equal to `n // i` (meaning `n` is not a perfect square, or `i` is not the square root), then `n // i` is also added to `sum_divisors`.

3.  **Final Check:**
    *   After checking all potential divisors, the function returns `True` if `sum_divisors` is equal to the original number `n`, indicating it's a perfect number; otherwise, it returns `False`.

7. Check **Strong number** (sum of factorial of digits).

```python
import math

def factorial(n):
    if n < 0:
        return 0 # Factorial is not defined for negative numbers, returning 0 or raising error is common.
    if n == 0:
        return 1
    res = 1
    for i in range(1, n + 1):
        res *= i
    return res

def is_strong_number(n):
    if n < 0:
        return False # Strong numbers are typically positive integers

    original_n = n
    sum_of_factorials = 0
    temp_n = n

    while temp_n > 0:
        digit = temp_n % 10  # Get the last digit
        sum_of_factorials += factorial(digit) # Add factorial of the digit to the sum
        temp_n //= 10  # Remove the last digit

    return sum_of_factorials == original_n

# Test cases
number1 = 145 # 1! + 4! + 5! = 1 + 24 + 120 = 145
print(f"Is {number1} a strong number? {is_strong_number(number1)}")

number2 = 2 # 2! = 2
print(f"Is {number2} a strong number? {is_strong_number(number2)}")

number3 = 1 # 1! = 1
print(f"Is {number3} a strong number? {is_strong_number(number3)}")

number4 = 40585 # 4! + 0! + 5! + 8! + 5! = 24 + 1 + 120 + 40320 + 120 = 40585
print(f"Is {number4} a strong number? {is_strong_number(number4)}")

number5 = 123
print(f"Is {number5} a strong number? {is_strong_number(number5)}")

number6 = 0 # Conventionally, 0 is not a strong number, or treated as False.
print(f"Is {number6} a strong number? {is_strong_number(number6)}")

number7 = -145
print(f"Is {number7} a strong number? {is_strong_number(number7)}")
```

### How the code works:

1.  **`factorial(n)` function (Helper):**
    *   This function calculates the factorial of a given non-negative integer `n`.
    *   It returns `1` for `n = 0` (by definition) and handles negative inputs by returning `0` (or it could raise an error).

2.  **`is_strong_number(n)` function:**
    *   It takes an integer `n` as input.
    *   **Edge Cases:**
        *   Returns `False` for `n < 0` as strong numbers are typically positive.
        *   `original_n` stores the initial value of `n` for later comparison.
    *   `sum_of_factorials` is initialized to `0`.
    *   `temp_n` is used to extract digits without modifying the `original_n`.

3.  **Calculating Sum of Factorials:**
    *   A `while temp_n > 0:` loop extracts each `digit` using the modulo operator (`% 10`).
    *   For each `digit`, its factorial is calculated using the `factorial()` helper function and added to `sum_of_factorials`.
    *   `temp_n` is then floor-divided by 10 (`temp_n //= 10`) to remove the last digit.

4.  **Final Check:**
    *   After processing all digits, the function returns `True` if `sum_of_factorials` is equal to `original_n`; otherwise, it returns `False`.

8. Check **Harshad/Niven number**.

```python
def sum_digits_for_harshad(n):
    # Helper function to calculate sum of digits (reusing logic)
    temp_n = abs(n) # Harshad numbers are usually positive, but sum of digits can be for magnitude
    total_sum = 0
    while temp_n > 0:
        digit = temp_n % 10
        total_sum += digit
        temp_n //= 10
    return total_sum

def is_harshad_number(n):
    if n <= 0:
        return False # Harshad numbers are typically positive integers

    sum_of_digits = sum_digits_for_harshad(n)

    # A Harshad number is divisible by the sum of its digits
    return n % sum_of_digits == 0

# Test cases
number1 = 18 # 1+8 = 9, 18 % 9 == 0
print(f"Is {number1} a Harshad number? {is_harshad_number(number1)}")

number2 = 21 # 2+1 = 3, 21 % 3 == 0
print(f"Is {number2} a Harshad number? {is_harshad_number(number2)}")

number3 = 153 # 1+5+3 = 9, 153 % 9 == 0
print(f"Is {number3} a Harshad number? {is_harshad_number(number3)}")

number4 = 17 # 1+7 = 8, 17 % 8 != 0
print(f"Is {number4} a Harshad number? {is_harshad_number(number4)}")

number5 = 0
print(f"Is {number5} a Harshad number? {is_harshad_number(number5)}")

number6 = -12
print(f"Is {number6} a Harshad number? {is_harshad_number(number6)}")

number7 = 100 # 1+0+0 = 1, 100 % 1 == 0
print(f"Is {number7} a Harshad number? {is_harshad_number(number7)}")
```

9. Check **Automorphic numbe**r (ends with its square).

```python
import math

def is_automorphic_number(n):
    if n < 0:
        return False # Automorphic numbers are typically non-negative
    if n == 0:
        return True # 0*0 = 0, ends with 0
    if n == 1:
        return True # 1*1 = 1, ends with 1

    square = n * n

    # Determine the number of digits in n to extract the correct suffix from its square
    # This can be done by taking log10(n) + 1, or repeatedly dividing by 10
    num_digits_n = 0
    temp_n = n
    while temp_n > 0:
        temp_n //= 10
        num_digits_n += 1

    # Calculate the power of 10 needed to get the last 'num_digits_n' digits
    divisor = 10 ** num_digits_n

    # Check if the last 'num_digits_n' digits of the square are equal to n
    return square % divisor == n

# Test cases
number1 = 1 # 1*1 = 1
print(f"Is {number1} an automorphic number? {is_automorphic_number(number1)}")

number2 = 5 # 5*5 = 25
print(f"Is {number2} an automorphic number? {is_automorphic_number(number2)}")

number3 = 6 # 6*6 = 36
print(f"Is {number3} an automorphic number? {is_automorphic_number(number3)}")

number4 = 25 # 25*25 = 625
print(f"Is {number4} an automorphic number? {is_automorphic_number(number4)}")

number5 = 76 # 76*76 = 5776
print(f"Is {number5} an automorphic number? {is_automorphic_number(number5)}")

number6 = 376 # 376*376 = 141376
print(f"Is {number6} an automorphic number? {is_automorphic_number(number6)}")

number7 = 123 # 123*123 = 15129
print(f"Is {number7} an automorphic number? {is_automorphic_number(number7)}")

number8 = 0 # 0*0 = 0
print(f"Is {number8} an automorphic number? {is_automorphic_number(number8)}")

number9 = -5
print(f"Is {number9} an automorphic number? {is_automorphic_number(number9)}")
```

    *   `square` calculates the square of the input number `n`.

2.  **Counting Digits:**
    *   `num_digits_n` counts the number of digits in the original number `n`. This is crucial because we need to check if the square ends with *exactly* that many digits matching `n`.
    *   A `while temp_n > 0:` loop repeatedly divides `temp_n` by 10 (`temp_n //= 10`) and increments `num_digits_n` until `temp_n` becomes 0.

3.  **Extracting Trailing Digits and Final Check:**
    *   `divisor = 10 ** num_digits_n` calculates the appropriate power of 10. For example, if `n` has 2 digits, `divisor` will be 100.

10. Check **Neon number** (sum digits of square equals number).

```python
def sum_digits(num):
    # Helper function to calculate sum of digits
    total_sum = 0
    if num < 0:
        num = abs(num) # Handle negative numbers for sum of digits calculation
    if num == 0:
        return 0
    while num > 0:
        total_sum += num % 10
        num //= 10
    return total_sum

def is_neon_number(n):
    if n < 0:
        return False # Neon numbers are typically non-negative

    square = n * n
    sum_of_square_digits = sum_digits(square)

    return sum_of_square_digits == n

# Test cases
number1 = 0   # 0*0 = 0, sum of digits(0) = 0. 0 == 0 -> True
print(f"Is {number1} a Neon number? {is_neon_number(number1)}")

number2 = 1   # 1*1 = 1, sum of digits(1) = 1. 1 == 1 -> True
print(f"Is {number2} a Neon number? {is_neon_number(number2)}")

number3 = 9   # 9*9 = 81, sum of digits(81) = 8+1 = 9. 9 == 9 -> True
print(f"Is {number3} a Neon number? {is_neon_number(number3)}")

number4 = 45  # 45*45 = 2025, sum of digits(2025) = 2+0+2+5 = 9. 9 != 45 -> False
print(f"Is {number4} a Neon number? {is_neon_number(number4)}")

number5 = 123 # 123*123 = 15129, sum of digits(15129) = 1+5+1+2+9 = 18. 18 != 123 -> False
print(f"Is {number5} a Neon number? {is_neon_number(number5)}")

number6 = -9 # Negative numbers are not neon numbers
print(f"Is {number6} a Neon number? {is_neon_number(number6)}")
```

### How the code works:

    *   It uses a `while` loop with modulo (`% 10`) to extract the last digit and floor division (`// 10`) to remove it, accumulating the sum.

2.  **`is_neon_number(n)` function:**

```python
def gcd_euclidean(a, b):
    # The Euclidean algorithm states that gcd(a, b) = gcd(b, a % b)
    # and gcd(a, 0) = a.
    # We need to ensure that 'a' and 'b' are non-negative for standard GCD.
    a = abs(a)
    b = abs(b)

    while b:
        a, b = b, a % b
    return a

# Test cases
number_a1, number_b1 = 48, 18
print(f"The GCD of {number_a1} and {number_b1} is: {gcd_euclidean(number_a1, number_b1)}")

number_a2, number_b2 = 101, 103
print(f"The GCD of {number_a2} and {number_b2} is: {gcd_euclidean(number_a2, number_b2)}")

number_a3, number_b3 = 0, 5
print(f"The GCD of {number_a3} and {number_b3} is: {gcd_euclidean(number_a3, number_b3)}")

number_a4, number_b4 = 21, 0
print(f"The GCD of {number_a4} and {number_b4} is: {gcd_euclidean(number_a4, number_b4)}")

number_a5, number_b5 = -15, 25
print(f"The GCD of {number_a5} and {number_b5} is: {gcd_euclidean(number_a5, number_b5)}")

number_a6, number_b6 = 17, 17
print(f"The GCD of {number_a6} and {number_b6} is: {gcd_euclidean(number_a6, number_b6)}")
```

### How the code works:

1.  **`gcd_euclidean(a, b)` function:**
    *   It takes two integers, `a` and `b`, as input.
    *   **Absolute Values:** It first converts both `a` and `b` to their absolute values (`abs(a)`, `abs(b)`). The GCD is typically defined for non-negative integers, and this ensures correct behavior for negative inputs.

2.  **Euclidean Algorithm (`while b:` loop):**
    *   The core of the function is a `while` loop that continues as long as `b` is not zero.
    *   In each iteration, the values of `a` and `b` are updated: the new `a` becomes the current `b`, and the new `b` becomes the remainder of `a` divided by `b` (`a % b`).
    *   This process is based on the Euclidean algorithm's property: `gcd(a, b) = gcd(b, a % b)`.

3.  **Return Value:**
    *   When `b` becomes `0`, the loop terminates. At this point, `a` holds the GCD of the original two numbers. This is because `gcd(a, 0) = a`.
    *   The function then returns `a`.

12. Find **LCM** using GCD.

```python
def gcd_euclidean(a, b):
    # The Euclidean algorithm states that gcd(a, b) = gcd(b, a % b)
    # and gcd(a, 0) = a.
    # We need to ensure that 'a' and 'b' are non-negative for standard GCD.
    a = abs(a)
    b = abs(b)

    while b:
        a, b = b, a % b
    return a

def lcm_using_gcd(a, b):
    if a == 0 or b == 0:
        return 0 # LCM is 0 if either number is 0

    # LCM(a, b) = |a * b| / GCD(a, b)
    return abs(a * b) // gcd_euclidean(a, b)

# Test cases
number_a1, number_b1 = 4, 6
print(f"The LCM of {number_a1} and {number_b1} is: {lcm_using_gcd(number_a1, number_b1)}")

number_a2, number_b2 = 12, 18
print(f"The LCM of {number_a2} and {number_b2} is: {lcm_using_gcd(number_a2, number_b2)}")

number_a3, number_b3 = 7, 5
print(f"The LCM of {number_a3} and {number_b3} is: {lcm_using_gcd(number_a3, number_b3)}")

number_a4, number_b4 = 0, 10
print(f"The LCM of {number_a4} and {number_b4} is: {lcm_using_gcd(number_a4, number_b4)}")

number_a5, number_b5 = -8, 12
print(f"The LCM of {number_a5} and {number_b5} is: {lcm_using_gcd(number_a5, number_b5)}")

number_a6, number_b6 = 9, 9
print(f"The LCM of {number_a6} and {number_b6} is: {lcm_using_gcd(number_a6, number_b6)}")
```

### How the code works:

1.  **`gcd_euclidean(a, b)` function:**
    *   This is the helper function for calculating the Greatest Common Divisor (GCD) using the Euclidean algorithm, as implemented previously.
    *   It ensures `a` and `b` are non-negative and iteratively finds the GCD.

2.  **`lcm_using_gcd(a, b)` function:**
    *   It takes two integers, `a` and `b`, as input.
    *   **Edge Case:** If either `a` or `b` is `0`, their LCM is `0`. This case is handled first.
    *   **Core Logic:** The fundamental relationship between LCM and GCD is `LCM(a, b) = (|a * b|) / GCD(a, b)`.
        *   It calculates the product of the absolute values of `a` and `b` (`abs(a * b)`).
        *   It then divides this product by the GCD of `a` and `b`, obtained by calling the `gcd_euclidean` function.
        *   The `//` operator performs integer division, ensuring the result is an integer.

3.  **Return Value:**
    *   The function returns the calculated LCM.

13. Find **HCF** of three numbers.

```python
def gcd_euclidean(a, b):
    # Reusing the previously defined Euclidean algorithm for GCD
    a = abs(a)
    b = abs(b)
    while b:
        a, b = b, a % b
    return a

def hcf_three_numbers(a, b, c):
    # HCF of three numbers can be found by applying GCD iteratively:
    # HCF(a, b, c) = HCF(HCF(a, b), c)
    result = gcd_euclidean(a, b)
    result = gcd_euclidean(result, c)
    return result

# Test cases
number_a1, number_b1, number_c1 = 12, 18, 24
print(f"The HCF of {number_a1}, {number_b1}, and {number_c1} is: {hcf_three_numbers(number_a1, number_b1, number_c1)}")

number_a2, number_b2, number_c2 = 7, 13, 19 # Prime numbers
print(f"The HCF of {number_a2}, {number_b2}, and {number_c2} is: {hcf_three_numbers(number_a2, number_b2, number_c2)}")

number_a3, number_b3, number_c3 = 30, 45, 60
print(f"The HCF of {number_a3}, {number_b3}, and {number_c3} is: {hcf_three_numbers(number_a3, number_b3, number_c3)}")

number_a4, number_b4, number_c4 = 0, 10, 15
print(f"The HCF of {number_a4}, {number_b4}, and {number_c4} is: {hcf_three_numbers(number_a4, number_b4, number_c4)}")

number_a5, number_b5, number_c5 = -12, 18, 30
print(f"The HCF of {number_a5}, {number_b5}, and {number_c5} is: {hcf_three_numbers(number_a5, number_b5, number_c5)}")
```

    *   It handles non-negative integers by converting inputs to their absolute values.

2.  **`hcf_three_numbers(a, b, c)` function:**
    *   It takes three integers, `a`, `b`, and `c`, as input.
    *   **Core Logic:** The Highest Common Factor (HCF) of three numbers is found by applying the GCD operation iteratively.
        *   First, it calculates `result = gcd_euclidean(a, b)`, finding the HCF of the first two numbers.
        *   Then, it calculates the HCF of this `result` and the third number `c` (`result = gcd_euclidean(result, c)`).

3.  **Return Value:**

14. Compute **power** a^b using loop (no **).

```python
def power(base, exponent):
    if exponent == 0:
        return 1  # Any number raised to the power of 0 is 1

    result = 1.0
    if exponent > 0:
        for _ in range(exponent):
            result *= base
    else:  # exponent < 0
        # For negative exponents, calculate 1 / (base ^ |exponent|)
        positive_exponent = abs(exponent)
        for _ in range(positive_exponent):
            result *= base
        result = 1.0 / result
    return result

# Test cases
base1, exp1 = 2, 3   # 2^3 = 8
print(f"{base1}^{exp1} = {power(base1, exp1)}")

base2, exp2 = 5, 0   # 5^0 = 1
print(f"{base2}^{exp2} = {power(base2, exp2)}")

base3, exp3 = 10, 1  # 10^1 = 10
print(f"{base3}^{exp3} = {power(base3, exp3)}")

base4, exp4 = 2, -2  # 2^-2 = 1/4 = 0.25
print(f"{base4}^{exp4} = {power(base4, exp4)}")

base5, exp5 = -3, 2  # (-3)^2 = 9
print(f"{base5}^{exp5} = {power(base5, exp5)}")

base6, exp6 = -3, 3  # (-3)^3 = -27
print(f"{base6}^{exp6} = {power(base6, exp6)}")

base7, exp7 = 0, 5   # 0^5 = 0
print(f"{base7}^{exp7} = {power(base7, exp7)}")
```

### How the code works:

1.  **`power(base, exponent)` function:**
    *   It takes two numbers, `base` and `exponent`, as input.

2.  **Edge Cases:**
    *   **`exponent == 0`**: Any non-zero number raised to the power of 0 is 1. This is handled by immediately returning `1`.
    *   `result` is initialized to `1.0` to handle potential fractional results from negative exponents.

3.  **Positive Exponents (`exponent > 0`):**
    *   A `for` loop iterates `exponent` times.
    *   In each iteration, `result` is multiplied by the `base`. This performs repeated multiplication (e.g., `base * base * base` for `exponent = 3`).

4.  **Negative Exponents (`exponent < 0`):**
    *   The property `a^-b = 1 / (a^b)` is used.
    *   The absolute value of the `exponent` is taken (`positive_exponent = abs(exponent)`).
    *   A `for` loop then calculates `base` raised to `positive_exponent` (just like the positive exponent case).
    *   Finally, `1.0` is divided by this calculated `result` to get the correct value for the negative exponent.

5.  **Return Value:**
    *   The function returns the calculated `result` (the `base` raised to the `exponent`).

15. Compute power a^b mod m (fast exponentiation).

```python
def power_mod(base, exponent, modulus):
    if modulus == 1:
        return 0 # Any number modulo 1 is 0

    res = 1 # Initialize result
    base %= modulus # Update base to be base % modulus

    while exponent > 0:
        # If exponent is odd, multiply base with result
        if exponent % 2 == 1:
            res = (res * base) % modulus

        # exponent must be even now
        base = (base * base) % modulus
        exponent //= 2 # exponent = exponent / 2

    return res

# Test cases
# Example 1: 2^3 mod 5 = 8 mod 5 = 3
base1, exp1, mod1 = 2, 3, 5
print(f"{base1}^{exp1} mod {mod1} = {power_mod(base1, exp1, mod1)}")

# Example 2: 7^10 mod 3 = 282475249 mod 3 = 1
base2, exp2, mod2 = 7, 10, 3
print(f"{base2}^{exp2} mod {mod2} = {power_mod(base2, exp2, mod2)}")

# Example 3: 10^0 mod 7 = 1 mod 7 = 1
base3, exp3, mod3 = 10, 0, 7
print(f"{base3}^{exp3} mod {mod3} = {power_mod(base3, exp3, mod3)}")

# Example 4: Large exponent 3^100 mod 13
# 3^100 mod 13 = ((3^2)^50) mod 13 = (9^50) mod 13
# (9^2)^25 mod 13 = (81^25) mod 13 = (3^25) mod 13
# ... this continues efficiently
base4, exp4, mod4 = 3, 100, 13
print(f"{base4}^{exp4} mod {mod4} = {power_mod(base4, exp4, mod4)}")

# Example 5: Base is greater than modulus
base5, exp5, mod5 = 15, 2, 7 # 15^2 mod 7 = 225 mod 7 = 1
print(f"{base5}^{exp5} mod {mod5} = {power_mod(base5, exp5, mod5)}")
```

### How the code works (Fast Exponentiation / Modular Exponentiation):

The `power_mod(base, exponent, modulus)` function computes `(base ^ exponent) % modulus` efficiently. This algorithm is based on the idea that any exponent can be written in binary form.

1.  **Initialization:**
    *   `if modulus == 1`: If the modulus is 1, the result of any modulo operation is 0. This is an edge case handled first.
    *   `res = 1`: The result is initialized to 1. This is the neutral element for multiplication.
    *   `base %= modulus`: The `base` is reduced modulo `modulus`. This is important because `(a * b) % m = ((a % m) * (b % m)) % m`. Reducing the base early helps keep intermediate calculations smaller.

2.  **Iterative Calculation (`while exponent > 0` loop):**
    *   The loop continues as long as `exponent` is greater than 0.
    *   **Checking the LSB (Least Significant Bit) of the exponent:**
        *   `if exponent % 2 == 1`: If the current `exponent` is odd (meaning its least significant bit is 1), it implies that this power of the base contributes to the final product. So, `res` is updated by multiplying it with the current `base` and taking the result modulo `modulus` (`res = (res * base) % modulus`).
    *   **Squaring the Base:**
        *   `base = (base * base) % modulus`: The `base` is squared and reduced modulo `modulus`. This prepares the `base` for the next higher power (e.g., if `base` was `x`, it becomes `x^2`, then `x^4`, `x^8`, etc., effectively mirroring the powers of 2 in the exponent's binary representation).
    *   **Halving the Exponent:**
        *   `exponent //= 2`: The `exponent` is integer-divided by 2. This is equivalent to shifting the binary representation of the exponent one bit to the right, moving to the next bit (next power of 2).

3.  **Return Value:**
    *   Once the `exponent` becomes 0, the loop terminates, and `res` holds the final calculated value of `(original_base ^ original_exponent) % modulus`.
